```python
# iniciando a analise do crescimento diário

# Agrupando por país/região e somando os casos confirmados
regiao_data = data.groupby('Country/Region').sum().reset_index()

# Removendo colunas desnecessárias
regiao_data = regiao_data.drop(columns=['Province/State', 'Lat', 'Long'])

# Transpondo o dataframe para ter datas nas colunas e países nas linhas
df_transposed = regiao_data.set_index('Country/Region').T

# Renomeando a coluna de índice para 'Date'
df_transposed.index.name = 'Date'

# Convertendo o índice para datetime
df_transposed.index = pd.to_datetime(df_transposed.index, format='%m/%d/%y')

# Escolhendo os países para calcular a taxa de crescimento diário
countries = ['Australia', 'Canada', 'China', 'Denmark', 'France', 'Netherlands', 'United Kingdom']

# Inicializando um DataFrame para armazenar as taxas de crescimento diário
growth_rates = pd.DataFrame()

# Calculando e armazenando a taxa de crescimento diário para cada país
for country in countries:
    growth_rates[f'{country}_growth_rate'] = df_transposed[country].pct_change() * 100

# Plotando a taxa de crescimento diário para cada país
plt.figure(figsize=(14, 7))
for country in countries:
    sns.lineplot(data=growth_rates[f'{country}_growth_rate'], label=country)

plt.title('Taxa de Crescimento Diária dos Casos Confirmados de COVID-19 em Diversos Países')
plt.xlabel('Data')
plt.ylabel('Taxa de Crescimento (%)')
plt.xticks(rotation=45)
plt.legend(title='País')
plt.grid(True)
plt.savefig("crescimento.png")
plt.show()
```

```python
# iniciando o estudo do aumento de casos com o tempo
# agrupando por pais e somando casos confirmados
regiao_data = data.groupby('Country/Region').sum().reset_index()

# removendo algumas colunas desnecessárias
regiao_data = regiao_data.drop(columns=['Province/State', 'Lat', 'Long'])

# transpondo dataframe para ter data nas colunas e paises nas linhas
df_transposed = regiao_data.set_index('Country/Region').T
df_transposed.index.name = 'Date'
df_transposed.index = pd.to_datetime(df_transposed.index, format='%m/%d/%y')

# selecionando paises para a analised
countries = ['Australia', 'Canada', 'China', 'Denmark', 'France', 'Netherlands', 'United Kingdom']

#iniciando df para armazenar o aumento dos casos
increase_rates = pd.DataFrame()

# Calculando e Armazenando o Aumento de Casos para Cada País
for country in countries:
    increase_rates[f'{country}_increase'] = df_transposed[country].diff().fillna(0)

# plotando o aumento em cada pais
plt.figure(figsize=(14, 7))
for country in countries:
    sns.lineplot(data=increase_rates[f'{country}_increase'], label=country)

plt.title('Aumento de Casos Confirmados de COVID-19 por Região ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Aumento de Casos')
plt.xticks(rotation=45)
plt.legend(title='País')
plt.grid(True)
plt.savefig("casos.png")
plt.show()
```
